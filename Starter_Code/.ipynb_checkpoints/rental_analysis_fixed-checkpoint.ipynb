## Load Data

## Dwelling Types Per Year

In this section, you will calculate the number of dwelling types per year. Visualize the results using bar charts and the Pandas plot function. 

**Hint:** Use the Pandas `groupby` function.

**Optional challenge:** Plot each bar chart in a different color.

```python
# Calculate the sum number of dwelling types units per year (hint: use groupby)
to_data_per_year = to_data.groupby('year').sum().drop(columns=["average_house_value","shelter_costs_owned", "shelter_costs_rented"])
to_data_per_year
```

```python
# Save the dataframe as a csv file
to_data_per_year.to_csv("Data/total_number_of_dwelling_per_year.csv")
```

```python
# Extract data for each year
year_2001 = pd.DataFrame(to_data_per_year.iloc[0,:])
year_2001.columns=['number_of_dwelling']

year_2006 = pd.DataFrame(to_data_per_year.iloc[1,:])
year_2006.columns=['number_of_dwelling']

year_2011 = pd.DataFrame(to_data_per_year.iloc[2,:])
year_2011.columns=['number_of_dwelling']

year_2016 = pd.DataFrame(to_data_per_year.iloc[3,:])
year_2016.columns=['number_of_dwelling']
```

```python
# Helper create_bar_chart function
def create_bar_chart(data, title, xlabel, ylabel, color):
    plt.bar(x=data.index, height=data.number_of_dwelling, color=color, width = 0.6)
    plt.title(title,fontweight='bold')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=90)
```

```python
# Create a bar chart per year to show the number of dwelling types
# Bar chart for 2001
bar_chart_2001 = create_bar_chart(year_2001,'Dwelling Types in Toronto in 2001', '2001', 'Dwelling Type Units', 'red')
```

```python
# Bar chart for 2006
bar_chart_2006 = create_bar_chart(year_2006,'Dwelling Types in Toronto in 2006', '2006', 'Dwelling Type Units', "blue")
```

```python
# Bar chart for 2011
bar_chart_2011 = create_bar_chart(year_2011,'Dwelling Types in Toronto in 2011', '2011', 'Dwelling Type Units', 'orange')
```

```python
# Bar chart for 2016
bar_chart_2016 = create_bar_chart(year_2016,'Dwelling Types in Toronto in 2016', '2016', 'Dwelling Type Units', 'fuchsia')
```

## Average Monthly Shelter Costs in Toronto Per Year

In this section, you will calculate the average monthly shelter costs for owned and rented dwellings and the average house value for each year. Plot the results as a line chart.

**Optional challenge:** Plot each line chart in a different color.

```python
# Calculate the average monthly shelter costs for owned and rented dwellings
average_costs = to_data[["shelter_costs_owned", 'shelter_costs_rented']].groupby("year").mean()
average_costs.head()
```

```python
# Helper create_line_chart function
def create_line_chart(data, title, xlabel, ylabel, color):
    plt.plot(data, color=color)
    plt.title(title,fontweight='bold')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
# Create two line charts, one to plot the monthly shelter costs for owned dwelleing and other for rented dwellings per year

# Line chart for owned dwellings
line_chart_owned = create_line_chart(average_costs["shelter_costs_owned"],"Average Monthly Shelter Cost for Owned Dwelling in Toronto", "Year", "Avg Monthly Shelter Costs", "blue")
```

```python
# Line chart for rented dwellings
line_chart_rented = create_line_chart(average_costs["shelter_costs_rented"],"Average Monthly Shelter Cost for Rented Dwelling in Toronto", "Year", "Avg Monthly Shelter Costs", "orange")
```

## Average House Value by Neighbourhood

In this section, you will use `hvplot` to create an interactive visualization of the average house value with a dropdown selector for the neighbourhood.

**Hint:** It will be easier to create a new DataFrame from grouping the data and calculating the mean house values for each year and neighbourhood.

```python
# Create a new DataFrame for average house value
df_avg_house_values = pd.DataFrame(to_data,columns=["neighbourhood", "average_house_value"]).groupby(["year","neighbourhood"]).mean().reset_index()
df_avg_house_values.head()
```

```python
# Use hvplot to create an interactive line chart of the average house value per neighbourhood
# The plot should have a dropdown selector for the neighbourhood
plot_avg_house_value = df_avg_house_values.hvplot(y="average_house_value", x="year", groupby="neighbourhood",yformatter="%.0f", ylabel="Avg. House Value")
plot_avg_house_value
```

## Neighbourhood Map

In this section, you will read in neighbourhoods location data and build an interactive map with the average house value per neighbourhood. Use a `scatter_mapbox` from Plotly express to create the visualization. Remember, you will need your Mapbox API key for this.

### Load Location Data

```python
# Load neighbourhoods coordinates data
file_path = Path("Data/toronto_neighbourhoods_coordinates.csv")
df_neighbourhood_locations = pd.read_csv(file_path)
df_neighbourhood_locations = df_neighbourhood_locations.set_index("neighbourhood")
df_neighbourhood_locations.head()
```

### Data Preparation

You will need to join the location data with the mean values per neighbourhood.

1. Calculate the mean values for each neighbourhood.

2. Join the average values with the neighbourhood locations.

```python
# Calculate the mean values for each neighborhood
# Comments: Mean values have been calculated above (df_avg_house_value)

# Join the average values with the neighbourhood locations
df_joint = pd.concat([df_neighbourhood_locations, df_avg_house_value],join="inner",axis=1).reset_index()
df_joint.head()
```

```python
# Checking for null
df_joint.isnull().sum()
```

```python
# Checking for duplicates
df_joint.duplicated().sum()
```

### Mapbox Visualization

Plot the average values per neighbourhood using a Plotly express `scatter_mapbox` visualization.

```python
# Create a scatter mapbox to analyze neighbourhood info
map_avg_values = px.scatter_mapbox(
    df_joint,
    lat="lat",
    lon="lon",
    color="average_house_value",
    size="average_house_value",
    color_continuous_scale=px.colors.sequential.Jet,
    zoom = 10,
    mapbox_style="open-street-map",
    height=500,
    title="Average House Values in Toronto"
)
map_avg_values.show()
```
