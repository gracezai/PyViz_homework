## Global available data

```python
# Getting the average house values for all and the top 10 expensive neighbourhoods
df_avg_house_values = pd.DataFrame(to_data,columns=["neighbourhood", "average_house_value"]).groupby(["year","neighbourhood"]).mean().reset_index()
df_avg_house_values = df_avg_house_values.set_index("year")

df_avg_house_by_neighbourhoods = pd.DataFrame(to_data,columns=["neighbourhood", "average_house_value"]).groupby(["neighbourhood"]).mean().reset_index()
df_avg_house_value_top10 = df_avg_house_by_neighbourhoods.sort_values(by="average_house_value", ascending=False).nlargest(10, "average_house_value")

# Calculate the mean number of dwelling types units per year
df_number_of_dwelling = pd.DataFrame(to_data).drop(columns=["average_house_value","shelter_costs_owned", "shelter_costs_rented"]).groupby(["year","neighbourhood"]).mean().reset_index()

# Calculate the average monthly shelter costs for owned and rented dwellings
average_costs = to_data[["shelter_costs_owned", 'shelter_costs_rented','average_house_value']].groupby("year").mean()

# Getting the data for Dwelling Types Per Year
to_data_per_year = to_data.groupby('year').sum().drop(columns=["average_house_value","shelter_costs_owned", "shelter_costs_rented"])
year_2001 = pd.DataFrame(to_data_per_year.iloc[0,:])
year_2006 = pd.DataFrame(to_data_per_year.iloc[1,:])
year_2011 = pd.DataFrame(to_data_per_year.iloc[2,:])
year_2016 = pd.DataFrame(to_data_per_year.iloc[3,:])

```

## Panel Visualization Functions

```python
# Define Panel visualization functions
def neighbourhood_map():
    map_avg_values = px.scatter_mapbox(
        df_map,
        lat="lat",
        lon="lon",
        color="average_house_value",
        size="average_house_value",
        color_continuous_scale=px.colors.sequential.Jet,
        zoom = 10,
        mapbox_style="open-street-map",
        height=500,
        title="Average House Values in Toronto"
    )
    return map_avg_values

def create_bar_2001():
    bar_plot = year_2001.hvplot.bar(
        xlabel='2001',
        ylabel='Dwelling Type Units',
        title='Dwelling Types in Toronto in 2001', 
        width=650, 
        height=500, 
        color='red',
        rot=90
    ).opts(yformatter='%.0f')
    return bar_plot

def create_bar_2006():
    bar_plot = year_2006.hvplot.bar(
        xlabel='2006',
        ylabel='Dwelling Type Units',
        title='Dwelling Types in Toronto in 2006', 
        width=650, 
        height=500, 
        color='blue',
        rot=90
    ).opts(yformatter='%.0f')
    return bar_plot

def create_bar_2011():
    bar_plot = year_2011.hvplot.bar(
        xlabel='2011',
        ylabel='Dwelling Type Units',
        title='Dwelling Types in Toronto in 2011', 
        width=650, 
        height=500, 
        color='orange',
        rot=90
    ).opts(yformatter='%.0f')
    return bar_plot

def create_bar_2016():
    bar_plot = year_2016.hvplot.bar(
        xlabel='2016',
        ylabel='Dwelling Type Units',
        title='Dwelling Types in Toronto in 2016', 
        width=650, 
        height=500, 
        color='fuchsia',
        rot=90
    ).opts(yformatter='%.0f')
    return bar_plot

def create_line_owned():
    line_plot = average_costs['shelter_costs_owned'].hvplot.line(
        xlabel="Year",
        ylabel="Avg Monthly Shelter Costs",
        title='Average Monthly Shelter Costs for Owned Dwellings in Toronto',
        color="blue"
    )
    return line_plot

def create_line_rented():
    line_plot = average_costs['shelter_costs_rented'].hvplot.line(
        xlabel="Year",
        ylabel="Avg Monthly Shelter Costs",
        title='Average Monthly Shelter Costs for Rented Dwellings in Toronto',
        color="orange"
    )
    return line_plot

def average_house_value():
    avg_house_plot = average_costs['average_house_value'].hvplot.line(
        xlabel="Year", 
        ylabel="Avg. House Value",
        title='Average House Value in Toronto', 
        color="blue"
    )
    return avg_house_plot

def average_value_by_neighbourhood():
    avg_by_neighbourhood_plot = df_avg_house_values.hvplot(
        groupby='neighbourhood',
        xlabel='Year',
        ylabel='Avg. House Value'
    ).opts(yformatter="%.0f")
    return avg_by_neighbourhood_plot

def number_dwelling_types():
    num_by_tpyes = df_number_of_dwelling.hvplot.bar(
        x="year", 
        groupby="neighbourhood",
        yformatter="%.0f", xlabel="Year", 
        ylabel="Dwelling Type Units", 
        rot=90, 
        height=500
    )
    return num_by_tpyes

def average_house_value_snapshot():
    avg_house_value_snapshot_plot = px.bar(
        df_avg_house_values,
        x='neighbourhood',
        y='average_house_value',
        facet_row=df_avg_house_values.index,
        title='Average House Values in Toronto per Neighbourhood',
        height=1000, 
        width=900, 
        hover_data=['average_house_value'],   
        color='average_house_value'
    )
    return avg_house_value_snapshot_plot

def top_most_expensive_neighbourhoods():
    avg_house_value_top10 = df_avg_house_value_top10.hvplot.bar(
        x="neighbourhood", 
        y="average_house_value",
        yformatter="%.0f", 
        xlabel="Neighbourhood", 
        ylabel="Avg. House Value", 
        rot=90, 
        height=500, 
        title="Top 10 Expensive Neighbourhoods in Toronto", 
        color="mediumblue"
    )
    return avg_house_value_top10

def sunburts_cost_analysis():
    """Sunburst chart to conduct a costs analysis of most expensive neighbourhoods in Toronto per year."""
    
    # YOUR CODE HERE!
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
dashboard_title ="# Real Estate Analysis of Toronto from 2001 to 2016"

# Define a welcome text
dashboard_welcome ="The dashboard presents a visual analysis of historical house values, dwelling types per neighbourhood and dwelling costs in Toronto, Ontario according to census data from 2001 to 2016."
welcome_plot = pn.Column(dashboard_title,dashboard_welcome, neighbourhood_map())


# Create a tab layout for the dashboard
dashboard_page = pn.Column(dashboard_title, dashboard_welcome, neighbourhood_map())
yearly_analysis = pn.Row(pn.Column(create_bar_2001(), create_bar_2011()),pn.Column(create_bar_2006(),create_bar_2016()))
shelter_home_value = pn.Column(create_line_owned(),create_line_rented(), average_house_value())
house_value_dwelling = pn.Row(pn.Column(average_value_by_neighbourhood(), number_dwelling_types()), average_house_value_snapshot())
expensive_neighbourhood = pn.Column(top_most_expensive_neighbourhoods())

# Create the main dashboard
dashboard_tabs = pn.Tabs(             
                ("Welcome", welcome_plot),
                ("Yearly Market Analysis", yearly_analysis),
                ("Shelter Costs vs Home Value", shelter_home_value),
                ("Neighbourhood Analysis", house_value_dwelling),
                ("Top Expensive Neighbourhoods", expensive_neighbourhood)
)
```
